```python


#Set project root variables
from pathlib import Path
import sqlite3
import pandas as pd

project_root = Path.cwd().parents[0]
db_path = r"C:\Users\palla\OneDrive\Documents\Coding Projects\FDA_FAERS\database\faers.db"

conn = sqlite3.connect(db_path)
```

```python

ROSOUV = pd.read_sql_query("""

SELECT DISTINCT drugname, COUNT(*) AS reports
FROM drug
WHERE drugname LIKE '%ROSUVASTATIN%'
GROUP BY drugname
ORDER BY reports DESC;
        """,
        conn)

ROSOUV.head(20)
```

```python
reactions = pd.read_sql_query("""
SELECT
    r.pt AS symptom,
    COUNT(*) AS symptom_count
FROM reac r
JOIN (
    SELECT DISTINCT primaryid 
    FROM drug
    WHERE drugname LIKE '%ROSUVASTATIN%'
) ros ON r.primaryid = ros.primaryid
GROUP BY r.pt
ORDER BY symptom_count DESC
""", conn)

reactions.head(20)
```
